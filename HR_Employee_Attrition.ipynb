```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("/content/WA_Fn-UseC_-HR-Employee-Attrition (1).csv")
print(data.head())
print(data.columns)
```

```python
#df.head
#Missing values -
print(data.isnull().sum())
```

```python
#attrition
print(data["Attrition"])
```

```python
data["Attrition_num"] =data["Attrition"].map({"Yes":1,"No":0})
```

```python
import numpy as np
print("Average Age:",np.mean(data["Age"]))
```

```python
print("Average Monthly Income:",np.mean(data["MonthlyIncome"]))
```

```python
dept_attrition =data.groupby("Department")["Attrition_num"].mean()
print(dept_attrition)
```

```python
#Average salary of each dept
jobrole_income = data.groupby("JobRole")["MonthlyIncome"].mean()
print(jobrole_income)
```

```python
sns.countplot(x= "Attrition" ,data=data)
plt.show()
```

```python
plt.hist(data["Age"],bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#salary vs attrition graph

sns.boxplot(x="Attrition",y="MonthlyIncome",data=data)
plt.title("Monthly Income vs Monthly Income")
plt.xlabel("Attribution")
plt.ylabel("Monthly Income")
```

```python
#heatmap
sns.heatmap(data.corr(numeric_only=True),annot=False)
plt.title("Correlation Heatmap")
```
